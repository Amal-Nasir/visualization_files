```python
import matplotlib.pyplot as plt

# Data for the graphs
categories1 = ['DGO', 'DGC', 'DGP', 'DGT', 'DGU']
categories2 = ['PGO', 'PGC', 'PGD', 'PGT', 'PGU']
values1 = [1.4, 4.6, 39.5, 40.5, 14.0]
values2 = [2.1, 4.2, 53.4, 1.7, 38.6]

# Create the figure and subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plot the bar graph in the first subplot
bars1 = ax1.bar(categories1, values1, width=0.4, align='center', color='gray')
ax1.set_ylim(0, 60)
#ax1.set_xlabel('Categories')
# Change the font size of the categories
ax1.set_xticklabels(categories1, fontsize=13)
ax1.set_ylabel('Percentage \n(Duration of the Behavior / Total Visit Duration)', fontsize=14)
ax1.set_title('Doctor-initiated Behaviors')

# Add value labels above each bar in the first subplot
for bar in bars1:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width() / 2, height+0.5,
             str(height), ha='center', va='bottom', fontsize=12)

# Plot the bar graph in the second subplot
bars2 = ax2.bar(categories2, values2, width=0.4, align='center', color='gray')
# Change the font size of the categories
ax2.set_xticklabels(categories2, fontsize=13)
ax2.set_ylim(0, 60)
#ax2.set_xlabel('Categories')
#ax2.set_ylabel('Values')
ax2.set_title('Patient-initiated Behaviors')

# Add value labels above each bar in the second subplot
for bar in bars2:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width() / 2, height+0.5,
             str(height), ha='center', va='bottom', fontsize=12)

# Adjust the spacing between subplots
plt.tight_layout()

# Display the graph
plt.show()
```
