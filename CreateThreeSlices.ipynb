```python
def above_and_below(max_slice_ID, new_df):
    
    #df_sorted = new_df.sort_values(axis = 0, ascending=True, ignore_index=True)
    df_sorted = new_df.sort_values(by=['imageZposition'], axis = 0, ascending=True, ignore_index=True)
    print("im here", df_sorted)
    #print("im here", max_slice_ID)
    
    for i in range(len(df_sorted)):    
        if  max_slice_ID == df_sorted.InstanceID[i]:
            #print(df_sorted.InstanceID[i]) # return the max_slice_ID.values e.g(118)
            #print('Max:', df_sorted.iloc[[i]])
            #large5 = data.nlargest(5, "Salary")
            Max_1 = df_sorted.iloc[[i]]
            #print("Max:", Max_1)
            #Max = df_sorted.nlargest(1, "Area")
            Max = Max_1.to_numpy()
            Max_lst.append(Max)
            
            if i < len(df_sorted) - 1:
                #print('above:',df_sorted.iloc[[i+1]])
                above = df_sorted.iloc[[i+1]]
                #if above != Max_1:
                #print("above:", above)
                above = above.to_numpy()
                above_lst.append(above)
             
            if i < len(df_sorted) - 2: 
                above_next = df_sorted.iloc[[i+2]]
                #if above_next != Max_1:
                #print("above_next:", above_next)
                above_next = above_next.to_numpy()
                above_next_lst.append(above_next)                
                
            if i > 0:
                #print('below:',df_sorted.iloc[[i-1]])
                below = df_sorted.iloc[[i-1]]
                #if below != Max_1:
                #print("below:", below)
                below = below.to_numpy()
                below_lst.append(below)
            
            if i > 1:
                below_next = df_sorted.iloc[[i-2]]
                #if below_next != Max_1:
                #print("below_next:", below_next)
                below_next = below_next.to_numpy()
                below_next_lst.append(below_next)

    return above_lst, above_next_lst, Max_lst, below_lst, below_next_lst  
```

```python
train_slices = pd.merge(select_slices, df_all_slices) #df_ID[df_ID["InstanceID"].isin(df_all_slices["InstanceID"])]        
display(train_slices)
```

```python
def get_max_info():
    
    for x in range(len(train_slices.index)): 
        
        max_slice_ID = train_slices["InstanceID"].iloc[x]
        rad_ID = train_slices["RadiologistID"].iloc[x]
        nod_ID = train_slices["noduleID"].iloc[x]

        # selecting max slice neighbors from the df_all_slices
        new_df = df_all_slices.loc[(df_all_slices["noduleID"] == nod_ID)& (df_all_slices["RadiologistID"] == rad_ID)]
        #print(new_df)
        # Adding the agreement for each slice
        new_df["Agreement"] = train_slices["Agreement_number"].iloc[x]
        new_df["Binary Rating"] = train_slices["Binary_Rating_number"].iloc[x]
        
        above, above_next, Max, below, below_next = above_and_below(max_slice_ID, new_df) #above,  Max, below
           
    return above, above_next, Max, below, below_next
```
